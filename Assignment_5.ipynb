```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

# Load the dataset

```python
data=pd.read_csv('ToyotaCorolla.csv')
data
```

# Select the features you want to normalize

```python
features = ['Price', 'Age_08_04', 'KM', 'HP', 'cc', 'Doors', 'Weight']
```

# Extract the selected features

```python
X = data[features]
```

# 1. Min-Max Scaling

```python
min_max_scaler = MinMaxScaler()
min_max_scaled = min_max_scaler.fit_transform(X)
```

# 2. Z-score Scaling

```python
z_score_scaler = StandardScaler()
z_score_scaled = z_score_scaler.fit_transform(X)
```

# 3. Decimal Scaling

```python
decimal_scaled = X.apply(lambda x: x / 10**(len(str(int(x.max()))) - 1))
```

# Convert the scaled data back to DataFrames for readability


```python
min_max_scaled_df = pd.DataFrame(min_max_scaled, columns=features)
z_score_scaled_df = pd.DataFrame(z_score_scaled, columns=features)
decimal_scaled_df = decimal_scaled
```

# Print the first few rows of each scaled dataset

```python
print("Min-Max Scaled Data:")
print(min_max_scaled_df.head())
```

```python
print("\nZ-score Scaled Data:")
print(z_score_scaled_df.head())
```

```python
print("\nDecimal Scaled Data:")
print(decimal_scaled_df.head())
```
